```python


# Cargar archivo
archivo = '/Users/admin/Documents/PYTHON_TESIS/NIST TABLAS/OxigenoNIST_ASD.csv'
df = pd.read_csv(archivo)

# Función para limpiar cada valor
def limpiar_valor(valor):
    if isinstance(valor, str):
        valor = valor.strip()
        if valor.startswith('="') and valor.endswith('"'):
            valor = valor[2:-1]
        elif valor.startswith('"') and valor.endswith('"'):
            valor = valor[1:-1]
        valor = valor.strip()
        try:
            return float(valor)
        except ValueError:
            return valor
    return valor

# Aplicar de forma moderna sin usar applymap (evita warning)
df = df.astype(str).apply(lambda col: col.map(limpiar_valor))

# Mostrar tipos y primeros datos
print("✅ Tipos de datos por columna:")
print(df.dtypes)
```

# ====== Código de OutputAE ======

```python
import pandas as pd

# Encabezados clave exactos
encabezados_clave = [
    "NL =  1  0   S.M.-POTENTIAL V(NL)",
    "GAM ( N,  L,NION, Z,SIG/D/P)",
    "I(R)   R(  A,  B,   C,  D, 2LBD )",
    "ABOVE THE DIAGONAL ARE VELOCITY INTEGRALS",
    "T,2S+1L  P    H(1)/2RY   EIGEN-H/2RY",
    "I         T        K*CM  2S+1 L   CF",
    "E1-DATA  I  IP      A(EK)*SEC",
    "2*RY                 1/CM             Z**4/CM",
    "GAM    N    L    E(NON.REL)",
    "I(R)  I(  A,  C ) =       MASS",
    "LV   T 2J 2S+1L  P      H(1)/2RY",
    "K        LV         T        K*CM",
    "I         T        K*CM  2S+1 L   CF",
    "E1-DATA   K  KP           A(EK)*SEC"
]

# Archivo a leer
archivo = '/Users/admin/Documents/PYTHON_TESIS/autoestructure/O/olg'

# Resultado final: diccionario con clave 'df1', 'df2', etc.
tablas = {}

# Lectura línea por línea
with open(archivo, 'r') as f:
    lineas = f.readlines()

i = 0  # índice de línea
contador = 1  # para df1, df2, etc.

while i < len(lineas):
    linea = lineas[i].strip()
    
    # ¿Es un encabezado de los clave?
    if any(linea.startswith(enc) for enc in encabezados_clave):
        encabezado_original = linea
        i += 1

        # Saltar líneas en blanco
        while i < len(lineas) and not lineas[i].strip():
            i += 1

        # Leer datos hasta encontrar línea vacía
        datos = []
        while i < len(lineas):
            fila = lineas[i].strip()
            if not fila:
                break
            datos.append(fila.split())
            i += 1

        if not datos:
            continue  # Saltar tablas vacías

        # Generar nombres de columnas genéricos: col1, col2, ..., colN
        n_columnas = len(datos[0])
        columnas = [f"col{j+1}" for j in range(n_columnas)]

        # Crear DataFrame con manejo de errores
        try:
            df = pd.DataFrame(datos, columns=columnas if all(len(f) == len(columnas) for f in datos) else None)
        except Exception as e:
            print(f"[⚠️] Error al construir DataFrame df{contador}: {e}")
            df = pd.DataFrame(datos)

        # Guardar en diccionario
        tablas[f'df{contador}'] = (encabezado_original, df)
        contador += 1

    else:
        i += 1  # avanzar a la siguiente línea

print(f"✅ Se detectaron {len(tablas)} tablas.")
```

```python
for nombre, (encabezado, df) in tablas.items():
    print("=" * 80)
    print(f"Nombre del DataFrame: {nombre}")
    print(f"Encabezado original: {encabezado}")
    print("\nNombres de columnas:")
    print(df.columns.tolist())

    # Intentar convertir columnas numéricas de forma segura
    for col in df.columns:
        try:
            df[col] = pd.to_numeric(df[col])
        except (ValueError, TypeError):
            continue  # Dejar como está si no se puede convertir

    print("\nPrimeras filas:")
    print(df.head())

    print("\nResumen de columnas:")
    info_df = pd.DataFrame({
        'Tipo de dato': df.dtypes,
        'Nulos': df.isnull().sum(),
        'No nulos': df.notnull().sum(),
        'Valores únicos': df.nunique()
    })
    print(info_df)
```
